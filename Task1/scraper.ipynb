```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_banks_in_Nepal"
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
```

```python
#printing all the html content of provided url
print(soup)
```

```python
#finding table by it's class name
table=soup.find('table', class_="wikitable sortable")
print(table)
```

```python
#alternative method to get table through index value
soup.find_all('table')[0]
```

```python
headers = table.find_all("th")
```

```python
headers
```

```python
header_list=[header.text.strip() for header in headers]
print(header_list)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=header_list)
df
```

```python
column_data = table.find_all("tr")
print(column_data)
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
df["Paid-up capital"] = df["Paid-up capital"].str.replace("रु","Rs")
```

```python
df
```

```python
df.to_csv(r"commercial_bank.csv", index=False)
```
